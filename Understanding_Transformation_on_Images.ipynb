```python
# RESIZING IMAGES
```

```python
print("Original image dimensions are : " , image.shape)
height , width , channels = image.shape
imshow(image)
```

```python
# changing the height and width to half

from skimage.transform import resize

resized = resize(image , (height/2 , width/2))
print("Resized image dimensions : " , resized.shape)
imshow(resized)
```

```python
# COLOR MANIPULATION
```

```python
from skimage import color

grayscale_image = color.rgb2gray(image)
grayscale_image.shape
```

```python
imshow(grayscale_image)
```

```python
image2 = color.gray2rgb(grayscale_image)

red_filter = [1,0,0]
imshow(image2*red_filter)
```

```python
image2 = color.gray2rgb(grayscale_image)
green_filter = [0,1,0]
imshow(image2*green_filter)
```

```python
image2 = color.gray2rgb(grayscale_image)
blue_filter = [0,0,1]
imshow(image2*blue_filter)
```

```python
image2 = color.gray2rgb(grayscale_image)
blue_green_filter = [0,1,1]
imshow(image2*blue_green_filter)
```

```python
# IMAGE ROTATION

from skimage.transform import rotate

print('Rotated image : ')
rotated = rotate(image , 270 , resize = True)
imshow(rotated)
```

```python
# IMAGE CROPPING
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

cropped1 = image[200:600 , 200:550]
cropped2 = image[250:550 , 40:240]

imshow(image)
plt.title('Original image')
```

```python
fig , ax = plt.subplots(nrows=1 , ncols=2 , figsize = (10,10))
ax[0].imshow(cropped1)
ax[0].set_title('Cropped Image 1')
ax[1].imshow(cropped2)
ax[1].set_title('Cropped Image 2')
```

```python
# FLIPPING THE IMAGES
```

```python
flipLR = np.fliplr(image)
plt.imshow(flipLR)
plt.title('Left to Right flipped')
```

```python
flipUD = np.flipud(image)
plt.imshow(flipUD)
print('Up to Down flipped image')
```

```python
# BRIGHTNESS MANIPULATION

from skimage.exposure import adjust_gamma

imshow(image)
print('Original image : ')
```

```python
bright = adjust_gamma(image , gamma=0.5 , gain = 1)
imshow(bright)
print('Brightened Image')
```

```python
dark = adjust_gamma(image , gamma = 1.5 , gain = 1)
imshow(dark)
print('Darkened Image : ')
```

```python
# SAVING THE IMAGES
```

```python
imsave('Understanding_transformations_on_Images/dark.jpg' , dark)

# go and check in the Understanding_transformations_on_Images folder in FILES.
```

```python
# HOW TO DEAL WITH MULTIPLE IMAGES

import glob

IMAGES = []
path = 'Understanding_transformations_on_Images/Image_features/*.jpg'

for img in glob.glob(path):
    IMAGES.append(imread(img))
```

```python
IMAGES = np.array(IMAGES)
IMAGES.shape
```

```python
# LET'S PLOT THE IMAGES

# We can also use matplotlib to visualize images, it gives greater control over 
# visualization
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

image_titles = ['Frame1' , 'Frame2' , 'Frame3']

# plotting multiple images using subplots

fig , ax = plt.subplots(nrows=3 , ncols=1 , figsize = (15,10))
for i in range(3):
    ax[i].imshow(IMAGES[i])
    ax[i].set_title(image_titles[i])
    ax[i].set_xticks([])
    ax[i].set_yticks([])
```
